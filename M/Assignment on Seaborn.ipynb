2.Write a Python program to create a graph to find relationship between the sepal length

```python
data=sns.load_dataset("iris")
sns.scatterplot(x="sepal_length",y="sepal_width",data=data)
plt.show()
```

3.Write a Python program to create a join plot using "hexbin" to describe individual
distributions on the same plot between Sepal length and Sepal width. (Using the dataset
“iris.csv”).

```python
import pandas as pd
iris=pd.read_csv("iris.csv")
fig=sns.jointplot(x="sepal.length",y="sepal.width",kind="hex",color="red",data=iris)
plt.show()
```

4.Write a Python program to create a joinplot and add regression and kernel density fits
using "reg" to describe individual distributions on the same plot between Sepal length
and Sepal width. (Using the dataset “iris.csv”).

```python
import pandas as pd
iris=pd.read_csv("iris.csv")
fig=sns.jointplot(x="sepal.length",y="sepal.width",kind="reg",color="red",data=iris)
plt.show()
```

5.Write a Python program to find the correlation between variables of iris data. Also create
a heatmap using Seaborn to present their relations (Using the dataset “iris.csv”).

```python
import numpy as np
import pandas as pd
iris=pd.read_csv("iris.csv")
x=iris.iloc[:,0:4]
f,ax=plt.subplots(figsize=(10,8))
corr=x.corr()
print(corr)
sns.heatmap(corr,mask=np.zeros_like(corr),cmap=sns.diverging_palette(220,10,as_cmap=True),square=True,ax=ax,linewidths=.5)
plt.show()
```
